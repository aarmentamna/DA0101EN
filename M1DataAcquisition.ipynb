# **Data Acquisition**

## Importacion de Librerias

The Pandas Library is a useful tool that enables us to read various datasets into a dataframe; our Jupyter notebook platforms have a built-in Pandas Library so that all we need to do is import Pandas without installing.

```python
import pandas as pd # IMPORTACION LIBRERIA PARA TRABAJAR CON PANDAS
import numpy as np # IMPORTACION DE LIBRERIA PARA TRABAJAR CON NUMPY
from google.colab import data_table # IMPORTACION DE LIBRERIA PARA DAR FORMATO A LOS DATAFRAMES 
data_table.enable_dataframe_formatter() # ENABLE DATAFRAME FORMATTER
```

```python
df = pd.read_csv('sample_data/auto.csv', header=None) # LECTURA DE ARCHIVO
data_table.DataTable(df,True,5,26,30) #FORMATO DEL DATAFRAME
```

# Manejando el DataFrame 

```python
data_table.DataTable(df.head(5)) # IMPRESION DE LAS 5 PRIMERAS LINEAS
```

```python
print("The last 10 rows of the dataframe\n")
data_table.DataTable(df.tail(10)) #IMPRESION DE LAS 10 PRIMERAS LINEAS
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers # AGREGAR HEADERS A LAS COLUMNAS
data_table.DataTable(df.head(10))
```

```python
df1=df.replace('?',np.NaN) # REMPLAZAR EL SIGNO ? POR NaN
data_table.DataTable(df1)
```

```python
df=df1.dropna(subset=["price"], axis=0) # ELIMINA LA COLUMNA PRECIO
data_table.DataTable(df.head(20))
```

```python
print(df.columns)
```

```python
df.to_csv("automobile.csv", index=False) # ALMACENAR EL DATAFRAME 
```
